## Write a Pandas program to implement following operation
* ### to create a dataframe from a dictionary and display it

```python
import numpy as np
import pandas as pd
d = {
    "name" : ['diya', 'riya', 'jiya', 'himani'],
    "en_no" : [7002, 7003, 7006, 7013],
    "age" : [18, 17, 17, 18]
}
df = pd.DataFrame(d)
print("DataFrame : \n", df)
```

* ### to sort the DataFrame first by 'name' in ascending order

```python
print(" Original DataFrame : \n", df)
df_new = df.sort_values(by = ['name'])
print("DataFrame sorted by name : \n", df_new)
```

* ### to delete the one specific column from the DataFrame

```python
print("DataFrame before deleting column : \n", df)
del df['age']
print("DataFrame after deleting 'age' column : \n", df)
```

* ### to write a DataFrame to CSV file using tab separator

```python
df.to_csv('/kaggle/working/dataset.csv', index=False, sep="\t")
dataset = pd.read_csv('/kaggle/working/dataset.csv')
print("Original DataFrame : \n", df)
print("Dataframe in csv seperated by tab : \n", dataset)
```
