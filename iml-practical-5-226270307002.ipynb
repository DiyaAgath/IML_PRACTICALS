## Write a NumPy program to implement following operation

* **to find the maximum and minimum value of a given flattened array**

```python
import numpy as np
```

```python
arr = np.array([[1 ,5 ,8], [3, 7, 9], [2, 4, 6]])
a = arr.flatten()
print("Array before flattening : \n", arr)
print("Array after flattening : \n",a)
```

```python
min = np.min(a)
max = np.max(a)

print("Minimum value in flattened array is : ", min)
print("Maximum value in flattened array is : ", max)
```

* ### **to compute the mean, standard deviation, and variance of a given array along the second axis**

```python
array = np.array([[55, 10, 28], [15, 20, 54], [70, 90, 82]])

print("Array : \n", array)

mean = np.mean(array, axis = 1)
std = np.std(array, axis = 1)
var = np.var(array, axis = 1)

print("Mean of array : ", mean)
print("Standard Deviation of array : ", std)
print("Variance of array : ", var)
```
