* to create another shape from an array without changing its data(3x2 to 2x3)

```python
d = np.array([[1,2],
             [3,4],
             [5,6]])
```

```python
print("array before reshape : \n",d)
```

```python
d = d.reshape(2,3)
```

```python
print("array after reshape : \n",d)
```
