## Write a Pandas program to implement following operation
* ### to convert a NumPy array to a Pandas series

```python
import numpy as np
import pandas as pd
```

```python
a=np.array([24,12,20,10])
print("Numpy Array :",a)

x=pd.Series(a)
print("Pandas Series :\n", x)
```

* ### to convert the first column of a DataFrame as a Series

```python
dict1 = {
    "name" : ['abc', 'pqr', 'xyz'],
    "marks" : [90, 95, 80],
    "city" : ['porbandar', 'porbandar', 'junagadh']
}
df = pd.DataFrame(dict1)
s = pd.Series(df['name'])
print("1st column(name) of dataframe as series : \n", s)
```

* ## to create the mean and standard deviation of the data of a given Series

```python
s1 = pd.Series([10, 26, 1, 7])
print("dataframe : \n", s1)
print("\nMean : ", s1.mean())
print("Standard Deviation : ", s1.std())
```

* ## to sort a given Series

```python
s2 = pd.Series([10, 52, 85, 74, 26, 2, 91, 125, 256, 68])
print("Series before sorting : \n", s2)
print("Series after sorting : \n", s2.sort_values())
```
