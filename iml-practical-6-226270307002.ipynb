## Write a Pandas program to implement following operation
* ### to convert a NumPy array to a Pandas series

```python
import numpy as np
import pandas as pd
```

```python
a=np.array([24,12,20,10])
print("Array :",a)

x=pd.Series(a)
print(" Series :\n", x)
```
